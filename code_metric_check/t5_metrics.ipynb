```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib_inline.backend_inline

matplotlib_inline.backend_inline.set_matplotlib_formats("png2x") # svg, retina, png2x ...
mpl.style.use("seaborn-v0_8")
mpl.rcParams.update({"figure.constrained_layout.use": True})
sns.set_context("paper") 
sns.set_palette("Set2") 
sns.set_style("whitegrid") 

plt.rc("font", family = "DejaVu Sans")
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd

val_loss_transfer_0 = pd.read_csv('t5/val_loss_transfer_0.csv')
val_loss_transfer_1 = pd.read_csv('t5/val_loss_transfer_1.csv')
val_loss_transfer_1_large = pd.read_csv('t5/val_loss_transfer_1_large.csv')
```

```python
val_loss_transfer_0
```

```python
fig, ax = plt.subplots(1,3,figsize = (15,5))

sns.lineplot(data=val_loss_transfer_0,
             x='Step',
             y='Value',
             ax=ax[0]);

sns.lineplot(data=val_loss_transfer_1,
             x='Step',
             y='Value',
             ax=ax[1]);

sns.lineplot(data=val_loss_transfer_1_large,
             x='Step',
             y='Value',
             ax=ax[2]);

ax[0].set_title('default')
ax[1].set_title('settings 1')
ax[2].set_title('settings 1 with colab(large)');
```
